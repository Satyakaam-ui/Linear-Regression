```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Salary_Data.csv")
F_data = data.iloc[:,:-1].values
T_data = data.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
F_data_train,F_data_test,T_data_train,T_data_test = train_test_split(F_data,T_data,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
Regressor = LinearRegression()
Regressor.fit(F_data_train,T_data_train)
```

```python
Pred_T_data = Regressor.predict(F_data_test)
print(Pred_T_data)
```

```python
print(T_data_test)
```

```python
plt.scatter(F_data_train,T_data_train,color = 'red')
plt.plot(F_data_train,Regressor.predict(F_data_train), color = 'blue')
plt.title('Salary vs Experience(Training Set)')
plt.xlabel('# Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(F_data_test,T_data_test,color = 'red')
plt.plot(F_data_test,Pred_T_data, color = 'blue')
plt.title('Salary vs Experience(Test Set)')
plt.xlabel('# Experience')
plt.ylabel('Salary')
plt.show()
```
